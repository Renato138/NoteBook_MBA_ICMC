### Questão 1 (2 pontos)
Considerando os dados contidos no arquivo <font style="font-family: monaco">covid.csv"</font>:


#### a) (0.5 pontos)
Carregue o arquivo "covid.csv" e armazena em um DataFrame. Imprima as 10 primeiras linhas do DataFrame.

```python
df = pd.read_csv("Data/covid.csv")
df.head(10)
```

#### b) (0.5 pontos)
Elimine a linha de índice 5 do DataFrame referente ao país ('Country/Region') `Antarctica`.

```python
df.drop(5, inplace=True)
df.head(10)
```

#### c) (1 ponto)
Remova a coluna `Province/State`.

```python
df.drop(columns="Province/State", inplace=True)
df.head()
```

### Questão 2 (2 pontos)
Remova do DataFrame todas as linhas cuja soma do total de casos seja menor que 10.000. Imprima quantas linhas foram removidos.

```python
print("Número de linhas atual no DataFrame: ", df.shape[0])
mask = df.iloc[:,3:-1].sum(axis=1) < 10000
print("Número de linhas a serem deletadas: ", mask.sum())
df.drop(df[mask].index, inplace=True)
print("Número de linhas após a deleção: ", df.shape[0])
df.head()
```

### Questão 3 (3 pontos)
Alguns países podem aparecer mais de uma vez no DataFrame, pois podem estar fragmentados em estados ou províncias. Sabendo disso, quais os 5 países que apresentaram o maior número de casos considerando a soma de todos os seus estados e províncias na data mais recente de registro (coluna `3/9/22`)?

```python
dfs = df.groupby("Country/Region")["3/9/22"].sum().sort_values(ascending=False)

#Nota: Usei a propriedade 'values' para não "poluir" a impressão
print("Países com maior número de casos: ", dfs.index[0:5].values)
```

### Questão 4 (3 pontos)
Faça o gráfico da evolução do número acumulado de contaminados no Brasil (Brazil) e na Itália (Italy) ao longo do tempo. 

```python
import matplotlib.pyplot as plt

datas = df.columns.values[3:-1]
casos_brasil = (df[df["Country/Region"] == 'Brazil'][datas]).to_numpy()[0]/1000000
casos_italia = (df[df["Country/Region"] == 'Italy'][datas]).to_numpy()[0]/1000000

plt.figure(figsize=(16,9))

plt.plot(casos_brasil)
plt.plot(casos_italia)

plt.legend(["Brasil","Itália"], loc='center left', bbox_to_anchor=(1, 0.5))

plt.title("Casos acumulados de covid no Brasil e na Itália")
plt.xlabel("Período de 22/Jan/20 a 09/Mar/22")
plt.ylabel("Nº casos (em milhões)")

ax = plt.gca()
ax.axes.xaxis.set_ticks([])

plt.show()
```
