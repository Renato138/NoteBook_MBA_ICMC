### Exercício 4) 
Encontre e imprima:
- os anos em que o Brasil ganhou a copa 
- os anos em que o Brasil jogou uma final (a final é dada pelas colunas 'Winner' e 'RunnersUp') 
- a proporção de vitórias do Brasil em finais

```python
# SOLUÇÃO
df_brasil_campeao = df.loc[df['Winner'] == 'Brazil']
df_brasil_campeao
```

```python
# SOLUÇÃO
df_brasil_final = df.loc[(df['Winner'] == 'Brazil') | (df['RunnersUp'] == 'Brazil')]
df_brasil_final
```

```python
print('Proporção de vitórias do Brasil em finais: ', df_brasil_campeao.shape[0], ' em ', df_brasil_final.shape[0], ' = ', df_brasil_campeao.shape[0]/df_brasil_final.shape[0])
```

### Exercício 8)
Construa um dicionário onde as chaves são os países que já participaram de uma final e os valores são o número de vezes que o país participou da final.

```python
# SOLUÇÃO
s1 = df.value_counts('Winner').add(df.value_counts('RunnersUp'), fill_value=0).add(df.value_counts('Third'), fill_value=0).add(df.value_counts('Fourth'), fill_value=0).to_dict()
print(s1)
```

### Exercício 9) 
Construa um dicionário com todos os países que já participaram de uma semi-final de copa (os quatro finalistas), onde a chave é o país e o valor corresponde ao número de vezes que o país foi um semi-finalista.

```python
# SOLUÇÃO
s1 = df.value_counts('Winner').add(df.value_counts('RunnersUp'), fill_value=0).add(df.value_counts('Third'), fill_value=0).add(df.value_counts('Fourth'), fill_value=0).to_dict()
print(s1)
```
